```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()

transform = encoder.fit_transform(df[['smoking_history']]).toarray()
labels = encoder.categories_

join_rows=np.concatenate(labels)
UpdatedFile = pd.DataFrame(data = transform, columns = join_rows)
UpdatedFile

OpenFile = pd.concat((UpdatedFile, df), axis = 1)
```

```python
encoder = OneHotEncoder()

transform2 = encoder.fit_transform(df[['gender']]).toarray()
labels = encoder.categories_

join_rows2=np.concatenate(labels)
UpdatedFile2 = pd.DataFrame(data = transform2, columns = join_rows2)
UpdatedFile2
```

```python
OpenFile = pd.concat((UpdatedFile2, OpenFile), axis = 1)
OpenFile.head()
```

```python
OpenFile = OpenFile.drop(['gender'], axis = 1)
OpenFile = OpenFile.drop(['smoking_history'], axis = 1)
```
